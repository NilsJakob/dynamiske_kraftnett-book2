# Application Programming Interface (API)

Med Application Programming Interface eller API, kan vi få tilgang til databased direkte gjennom en API-nøkkel (API-KEY).

API-nøkkelen er en adgangsnøkkel til data, som identifiserer hvem som henter data med en digital ID, og bestemmer hvilke og hvor mye data du får lov til å hente.

Mange tjenester gir deg en gratis API-nøkkel ved registrering, men ofte er tjenesten laget slik at dette kun gjelder ved et lavt antall kall f.eks. 100–10 000 kall per dag med begrenset hastighet (rate limits) og litt lavere datakvalitet / forsinkelse, enn de tjenestene man betaler for.

```python
from entsoe import EntsoePandasClient
import pandas as pd
import os
from dotenv import load_dotenv

# Load .env file
load_dotenv()

# Hent API key fra miljøvariabel
api_key = os.getenv("API_KEY")

client = EntsoePandasClient(api_key=api_key)

start = pd.Timestamp("2025-01-01", tz="Europe/Oslo")
end   = pd.Timestamp("2026-01-01", tz="Europe/Oslo")

data = client.query_load("NO", start=start, end=end)

print(data.head())
```
