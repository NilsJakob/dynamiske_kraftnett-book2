# Klassifisering av stabilitet

Stabiliteten i kraftsystemet klassifiseres i rotorvinkel‑, spennings‑, frekvens‑, omformerdrevet‑ og resonansstabilitet.
Vannkraftverk bidrar til robustheten i kraftnettet gjennom treghet (inerti) og aktiv effektkontroll, men også gjennom reaktiv effektkontroll fra eksitasjonssystemet.
Kraftsystemstabilitet kan forklares som nettets evne til å gjenopprette likevekt etter å ha blitt utsatt for en forstyrrelse.
Hvis dette ikke lykkes, kan det føre til en delvis eller total kollaps av systemet.
Konvensjonell kraftsystemstabilitet fokuserer på rotorvinkel, frekvens og spenning.
Generatorens bidrag til rotorvinkelstabilitet er dens evne til å opprettholde synkronisme mellom rotasjonshastigheten og systemfrekvensen etter en forstyrrelse.

Dette oppnås primært ved å opprettholde en balanse mellom den mekaniske tilført effekten (Pm) og den elektriske avgitt effekten (Pe), som beskrevet gjennom svingligningen (Swing equation), utledet fra Newtons andre lov for rotasjonsbevegelse 
$$
\frac{2H}{\omega_s} \frac{d^2 \delta}{dt^2} = P_m - P_e
$$
Her er H [J/MVA] treghetskonstanten, som beskriver den totale kinetiske energien lagret i generatoren.

### Frekvensstabilitet
Frekvensstabilitet refererer til kraftsystemets evne til å opprettholde en stabil frekvens etter å ha blitt utsatt for en ubalanse mellom produksjon og forbruk [26].
Dette håndteres hovedsakelig av regulatoren (governor), som registrerer endringer i nettet, uttrykt som avvik i rotorhastigheten, og automatisk justerer vanninntaket for å gjenopprette effektbalansen i svingligningen.
Denne responsen skjer vanligvis gjennom en droop‑regulering i regulatorsystemet.
Dersom en forstyrrelse oppstår, er det først maskinens treghetsrespons som virker. Deretter følger aktivering av droop‑regulering og til slutt automatisk generatorregulering (AGC).
Disse frekvensreservene bidrar til at frekvensen stabiliseres tilbake til sin nominelle verdi.

### Spenningsstabilitet
Spenningsstabilitet refererer til kraftsystemets evne til å opprettholde spenningen på et nominelt nivå i alle knutepunkter (busser) i systemet etter en forstyrrelse [26].
Dette kan styres ved hjelp av en automatisk spenningsregulator (AVR), som reagerer på ubalanser i reaktiv effekt som påvirker spenningsnivåene [30].
Generatoren kan regulere strømmen i eksitasjonssystemet, noe som endrer den reaktive effekten, og dermed justere driftspunktet for å støtte spenningen i nettet.

### Treghetskonstant
Et annet viktig aspekt er maskinens treghetskonstant. Ifølge svingligningen vil en høyere verdi av HHH føre til langsommere akselerasjon ved samme effektubalanse, noe som gjør systemet mindre følsomt for forstyrrelser.

### Resonans- og omformerdrevet stabilitet
Moderne stabilitetsfenomener (resonans og omformerdrevet stabilitet) er knyttet til effekten av hurtigvirkende kraftelektroniske (PE) enheter, som forventes å dominere kraftsystemet i fremtiden.
Resonans beskriver nettets evne til å opprettholde korrekt oscillasjon av spenning, strøm eller dreiemoment.
Omformerdrevet stabilitet beskriver nettets evne til å forbli stabilt når raske, programvarestyrte «beslutninger» fra kraftelektroniske omformere samspiller med de elektriske egenskapene i nettet.
